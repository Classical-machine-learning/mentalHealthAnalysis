# Cleaning words

```python
def cleaner(x):
    if len(x)<3:
        return False
    if not x.isalpha():
        return False
    x = x.strip()
    x = x.lower()
    return x

```

```python
cleaned = [cleaner(x) for x in set_words if cleaner(x)!=False]
```

# Finding in doc

```python
## Lemmatize doc
# preproc_doc = [ret_lem(x) for x in data.split()]
# preproc_doc = [x.lower().strip() for x in preproc_doc];preproc_doc
preproc_doc = [x.lower().strip() for x in data.split()];preproc_doc
```

```python
dic_tem = {}
```

```python
to_find = [x[0] for x in counts.most_common(60)];to_find
```

```python
for a in to_find:
    for b in range(1,len(preproc_doc)-2):
        if a==preproc_doc[b]:
            try:
                dic_tem[a].append(preproc_doc[b-1])
                dic_tem[a].append(preproc_doc[b+1])
                dic_tem[a].append(preproc_doc[b-1]+" "+preproc_doc[b]+" "+preproc_doc[b+1])
            except:
                dic_tem[a] = []
                dic_tem[a].append(preproc_doc[b-1])
                dic_tem[a].append(preproc_doc[b+1])
                dic_tem[a].append(preproc_doc[b-1]+" "+preproc_doc[b]+" "+preproc_doc[b+1])
    dic_tem[a] = list(set(dic_tem[a]))
```

```python
dic_tem.keys()
```

```python
for x in to_find:
    print(x+"~")
    test = ' '.join(dic_tem[x]).split(' ')
    test = [x for x in test if x not in stopwords]
    print(Counter(test).most_common(15))
```

# Text Generation

```python
import pandas as pd
from fastai.text import *
```

```python
data_new = open("data/scraped.txt",'r', encoding="utf8").read();data_new
```

```python
data_new = data_new.lower();data_new
```

```python
data_to_frame= pd.DataFrame([])
```

```python
for a in nltk.sent_tokenize(data_new):
    data_to_frame = data_to_frame.append([a])
```

```python
data_to_frame.head(10)
```

```python
data2 = (TextList.from_df(data_to_frame, cols=0)
                .split_by_rand_pct(0.1)
               .label_for_lm()  
                .databunch(bs=48))
```

```python
data2
```

```python
learn = language_model_learner(data2, AWD_LSTM, drop_mult=0.3)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(1, 1e-2, moms=(0.8,0.7))
```

```python
learn.fit_one_cycle(6, 1e-2, moms=(0.8,0.7))
```

```python
learn.save('stage-1')
```
